```python
def standardize_columns(df):
    df = df.copy()

    df.columns = (
        df.columns
        .str.strip()
        .str.lower()
    )

    rename_map = {
        "merchantname": "merchant_name",
        "merchanturl": "merchant_url",
        "provincename": "province_name",
        "regencyname": "regency_name",
        "districtname": "district_name",
    }

    df.rename(columns=rename_map, inplace=True)

    return df
```

```python
def normalize_text(text):

    if pd.isna(text):
        return ""

    text = str(text).lower()

    # remove non alphanumeric
    text = re.sub(r"[^a-z0-9\s]", " ", text)

    # remove multiple spaces
    text = re.sub(r"\s+", " ", text)

    return text.strip()
```

```python
TEXT_COLUMNS = [
    "merchant_name",
    "address",
    "province_name",
    "regency_name",
    "district_name",
]
```

```python
for sheet, df in dfs.items():

    for col in TEXT_COLUMNS:

        if col in df.columns:
            df[col] = df[col].apply(normalize_text)

    dfs[sheet] = df
```

```python
for sheet, df in dfs.items():

    merchant = df.get("merchant_name", "")
    address = df.get("address", "")

    df["combined_text"] = (
        merchant.fillna("") + " " +
        address.fillna("")
    )

    df["combined_text"] = (
        df["combined_text"]
        .apply(normalize_text)
    )

    dfs[sheet] = df
```

```python
from collections import Counter
```

```python
def extract_top_words(df, top_n=100):

    all_text = " ".join(
        df["combined_text"]
        .dropna()
        .astype(str)
        .tolist()
    )

    words = all_text.split()

    # filter kata terlalu pendek
    words = [
        word
        for word in words
        if len(word) > 3
    ]

    counter = Counter(words)

    return counter.most_common(top_n)
```

```python
for sheet, df in dfs.items():

    print("\n")
    print("=" * 70)
    print(f"SHEET: {sheet}")
    print("=" * 70)

    top_words = extract_top_words(df)

    for word, count in top_words:
        print(f"{word:<25} {count}")
```

```python
VALID_PHRASES = [
    "padang pariaman",
    "kab padang pariaman",
    "kota pariaman",
    "pariaman",
]

INVALID_PHRASES = [
    "lubuk begalung",
    "bandar buat",
    "kuranji",
    "ampang",
    "pauh",
    "parak",
    "andalas",
    "bukittinggi",
    "payakumbuh",
    "solok",
    "jakarta",
    "bekasi",
    "tangerang",
    "bogor",
    "bandung",
    "surabaya",
    "depok",
]
```

```python
def classify_location(text):

    text = normalize_text(text)

    # PRIORITY 1
    for phrase in VALID_PHRASES:
        if phrase in text:
            return "KEEP"

    # PRIORITY 2
    for phrase in INVALID_PHRASES:
        if phrase in text:
            return "REJECT"

    # PRIORITY 3
    return "REVIEW"
```

```python
for sheet, df in dfs.items():

    df["status"] = df["combined_text"].apply(
        classify_location
    )

    dfs[sheet] = df
```

```python
def classify_location_v2(text):

    text = normalize_text(text)

    # PRIORITY 1 — VALID
    for phrase in VALID_PHRASES:

        if phrase in text:

            return pd.Series({
                "status": "KEEP",
                "matched_rule": phrase,
                "confidence_score": 100
            })

    # PRIORITY 2 — INVALID
    for phrase in INVALID_PHRASES:

        if phrase in text:

            return pd.Series({
                "status": "REJECT",
                "matched_rule": phrase,
                "confidence_score": -100
            })

    # PRIORITY 3 — REVIEW
    return pd.Series({
        "status": "REVIEW",
        "matched_rule": "none",
        "confidence_score": 0
    })
```

```python
for sheet, df in dfs.items():

    # HAPUS kolom lama jika sudah ada
    cols_to_drop = [
        "status",
        "matched_rule",
        "confidence_score"
    ]

    existing_cols = [
        col for col in cols_to_drop
        if col in df.columns
    ]

    if existing_cols:
        df = df.drop(columns=existing_cols)

    # APPLY CLASSIFICATION
    classification_result = df[
        "combined_text"
    ].apply(classify_location_v2)

    # GABUNGKAN
    df = pd.concat(
        [df, classification_result],
        axis=1
    )

    dfs[sheet] = df
```

```python
summary_rows = []

for sheet, df in dfs.items():

    counts = (
        df["status"]
        .value_counts()
        .to_dict()
    )

    summary_rows.append({
        "sheet": sheet,
        "total_rows": len(df),
        "keep": counts.get("KEEP", 0),
        "review": counts.get("REVIEW", 0),
        "reject": counts.get("REJECT", 0),
    })

summary_df = pd.DataFrame(summary_rows)

summary_df
```

```python
OUTPUT_PATH = "../outputs/cleaned_result.xlsx"

with pd.ExcelWriter(
    OUTPUT_PATH,
    engine="openpyxl"
) as writer:

    # SUMMARY
    summary_df.to_excel(
        writer,
        sheet_name="SUMMARY",
        index=False
    )

    # EXPORT PER SHEET
    for sheet, df in dfs.items():

        df.to_excel(
            writer,
            sheet_name=sheet[:31],
            index=False
        )

        # KEEP ONLY
        keep_df = df[
            df["status"] == "KEEP"
        ]

        keep_df.to_excel(
            writer,
            sheet_name=f"{sheet[:20]}_KEEP",
            index=False
        )

        # REVIEW ONLY
        review_df = df[
            df["status"] == "REVIEW"
        ]

        review_df.to_excel(
            writer,
            sheet_name=f"{sheet[:20]}_REVIEW",
            index=False
        )

        # REJECT ONLY
        reject_df = df[
            df["status"] == "REJECT"
        ]

        reject_df.to_excel(
            writer,
            sheet_name=f"{sheet[:20]}_REJECT",
            index=False
        )

print("Export selesai!")
```
